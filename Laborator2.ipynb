Exercițiu laborator 2
Simularea variabilelor aleatoare

Considerăm o urnă în care avem 3 bile roșii, 4 albastre și 2 negre. Aruncăm un zar; dacă obținem un număr prim,
adăugăm o bilă neagră în urnă, dacă obținem 6, adăugăm o bilă roșie, iar în celelalte cazuri adăugăm o bilă albastră.
Apoi extragem o bilă din urnă.

a) Simulați în Python experimentul de mai sus.

b) Folosind modelarea făcută, estimați probabilitțtea de a obține o bilă roșie.
Bonus. Calculați probabilitatea teoretică a evenimentului de mai sus și comparați-o cu cea obținută.

```python
import random

urna_initiala = {'rosie': 3, 'albastra': 4, 'neagra': 2}

def actualizeaza_urna(urna):
    zar = random.randint(1, 6)

    if zar in [2, 3, 5]:
        urna['neagra'] += 1
    elif zar == 6:
        urna['rosie'] += 1
    else:
        urna['albastra'] += 1

def extrage_bila(urna):
    total_bile = sum(urna.values())
    lista_bile = ['rosie'] * urna['rosie'] + ['albastra'] * urna['albastra'] + ['neagra'] * urna['neagra']
    return random.choice(lista_bile)

def simuleaza_experimentul(numar_incercari):
    numar_rosii = 0

    for _ in range(numar_incercari):
        urna = urna_initiala.copy()
        actualizeaza_urna(urna)
        bila_extrasa = extrage_bila(urna)

        if bila_extrasa == 'rosie':
            numar_rosii += 1

    return numar_rosii / numar_incercari

probabilitate_estimata = simuleaza_experimentul(100000)
probabilitate_estimata
```
