```python
import networkx as nx
import matplotlib.pyplot as plt
from pgmpy.models import MarkovNetwork

model = MarkovNetwork()

nodes = ['A1', 'A2', 'A3', 'A4', 'A5']
edges = [('A1', 'A2'), ('A1', 'A3'), ('A2', 'A4'), ('A2', 'A5'), ('A3', 'A4'), ('A4', 'A5')]

model.add_edges_from(edges)

cliques = list(nx.find_cliques(model))

print("Clicile modelului Markov sunt:")
for clique in cliques:
    print(clique)

G = nx.Graph()
G.add_edges_from(edges)
pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=12, font_weight='bold')
plt.show()
```
